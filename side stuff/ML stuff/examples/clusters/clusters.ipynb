```python
import tensorflow as tf
from tensorflow import keras

import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('./data/train.csv')

color_map = {'red': 0, 'blue': 1, 'green': 2, 'teal': 3, 'orange': 4, 'purple': 5}
# color_map = {'red': [1, 0, 0, 0, 0, 0], 
#             'blue': [0, 1, 0, 0, 0, 0], 
#             'green': [0, 0, 1, 0, 0, 0], 
#             'teal': [0, 0, 0, 1, 0, 0], 
#             'orange': [0, 0, 0, 0, 1, 0],
#             'purple': [0, 0, 0, 0, 0, 1]
#             }
train_df["color"] = train_df.color.apply(lambda x: color_map[x])

np.random.shuffle(train_df.values)
```

```python
model = keras.Sequential([
	keras.layers.Dense(32, input_shape=(2,), activation='relu'),
	keras.layers.Dense(32, activation='relu'),
	# keras.layers.Dense(32, activation='relu'),
	# keras.layers.Dropout(0.2),
	# keras.layers.Dense(32, activation='relu'),
	# keras.layers.Dense(32, activation='relu'),
	# keras.layers.Dense(32, activation='relu'),
	keras.layers.Dense(6, activation='softmax')])
```

```python
model.compile(optimizer="adam", 
                #loss function is binary cross entropy, cuz we're doing binary classification
                #from_logits=True cuz our output is not normalized???
                loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True), 
                metrics=["accuracy"])
```

```python
attributes = np.column_stack((train_df.x.values, train_df.y.values))

model.fit(attributes, train_df.color.values, batch_size=4, epochs=10)
```

```python
test_df = pd.read_csv("./data/test.csv")
test_df["color"] = test_df.color.apply(lambda x: color_map[x])
test_x = np.column_stack((test_df.x.values, test_df.y.values))
```

```python
model.evaluate(test_x, test_df.color.values)
```
