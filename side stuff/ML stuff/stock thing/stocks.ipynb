```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
import pandas as pd
import yfinance as yf
import csv
```

```python
stock_list = pd.read_csv('nasdaq-listed_csv.csv')
symbols = stock_list['Symbol'].values
```

```python
for i in range(len(symbols)):
    
    try:
        df = yf.download(symbols[i], start='2019-01-01', end='2021-12-01')
        print(symbols[i])
        curStock = [symbols[i]]
        for index, row in df.iterrows():
            # print(row["Close"])
            curStock.append(row["Close"])
        # print("Last day: ", df.iloc[-1]["Close"])
        monthAhead = yf.download(symbols[i], start='2021-12-01', end='2021-12-26')
        # print("Next month:", monthAhead.iloc[-1]["Close"])
        monthDifference = ((monthAhead.iloc[-1]["Close"] / df.iloc[-1]["Close"]) - 1) * 100
        if (monthDifference > 30):
            curStock.append(1)
        else:
            curStock.append(0)

        # print(monthAhead["Close"])
        
        # uncomment later
        with open('stock_data.csv', 'a') as csvFile:
            writer = csv.writer(csvFile)
            writer.writerow(curStock)
    except:
        print('error')
```
